## Importing the libraries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

## Loading the image

```python
img = cv2.imread("C:/Users/venka/Downloads/image.jpg")
```

## Converting into RGB

```python
img_rgb=cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
```

```python
plt.title("CAT")
plt.imshow(img)
plt.show()
```

# Gray Color

```python
img_gray=cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.title("GRAY CAT")
plt.imshow(img_gray , cmap="gray")
plt.show()
```

# Inverted Image

```python
invert_img= cv2.bitwise_not(img_gray)
plt.title("Inverted Image")
plt.imshow(invert_img, cmap="gray")
plt.show()
```

# Blur Image

```python
blur_img=cv2.GaussianBlur(invert_img,(111,111),0)
plt.title("Blur Image")
plt.imshow(blur_img, cmap = "gray")
plt.show()
```

# Inverted Blur Image

```python
invert_blur=cv2.bitwise_not(blur_img)
plt.title("Inverted Blur Image")
plt.imshow(invert_blur,cmap="gray")
plt.show()
```

# Pencil Sketch Image

```python
sketch=cv2.divide(img_gray,invert_blur, scale =256.0)
plt.title("Sketch Image")
plt.imshow(sketch , cmap= "gray")
plt.show()
```
